# GFL ENVIRONMENTAL 2020 STOCK TREND ANALYSIS

```python
import warnings
warnings.filterwarnings('ignore')  # Hide warnings
import datetime as dt
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates
```

```python
#GFL 
#Getting stock price data

start = dt.datetime(2020, 1, 1)
end = dt.datetime.now()
df = web.DataReader("GFL", 'yahoo', start, end)  # Collects data
df.reset_index(inplace=True)
df.set_index("Date", inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(20, 12))
plt.plot(df[["High","Low","Open","Close"]])
plt.legend(labels=df[["High","Low","Open","Close"]])
plt.title('Stock Trend 2020', fontsize = 20)
plt.xlabel('Time')
plt.ylabel('Stock Value')
```

```python
# Create, and add 7 day moving average to dataframe

df_open_close= df.copy()
df_open_close['Close_7ma'] = df['Close'].rolling(window=7,min_periods=0).mean()
df_open_close['Open_7ma'] = df['Open'].rolling(window=7,min_periods=0).mean()

df_open_close.head()
```

```python
plt.figure(figsize=(20, 12))
plt.plot(df_open_close[["Close_7ma","Open_7ma","Open","Close"]])
plt.legend(labels=df_open_close[["Close_7ma","Open_7ma","Open","Close"]])
plt.title('Stock Trend Open Close 2020 7 Day MA', fontsize = 20)
plt.xlabel('Time')
plt.ylabel('Stock Value')
```

```python
#15days Moving Average

df_open_close['Close_15ma'] = df['Close'].rolling(window=15,min_periods=0).mean()
df_open_close['Open_15ma'] = df['Open'].rolling(window=15,min_periods=0).mean()

df_open_close.head()
```

```python
plt.figure(figsize=(20, 12))
plt.plot(df_open_close[["Close_15ma","Open_15ma","Open","Close"]])
plt.legend(labels=df_open_close[["Close_15ma","Open_15ma","Open","Close"]])
plt.title('Stock Trend Open Close 2020 15 Day MA', fontsize = 20)
plt.xlabel('Time')
plt.ylabel('Stock Value')
```

# TAKEAWAYS:-
1. Bull market refers to optimistic movement in stock market which means share prices rise, there is downfall in unemployment and economy is good whereas bear market refers to pessimistic movement in market which indicates that share price is falling, there is high unemployment and recession is approaching which means bull market is opposite to bear market.
2. A bear market is caused by a loss of investor, business, and consumer confidence. As confidence recedes, so does demand, and prices fall.
3. A bull market is a market financial situation which is characterized by the investor’s confidence, optimism and positive expectations that good results will continue.
4. The stocks of GFL Envvironmental saw a sharp decline during the onset of the lockdown due to COVID-19 from 2020-04-03. But it gradually increased from 2020-04-29.
5. The analysis shows that the stock trends are increasing in nature.
